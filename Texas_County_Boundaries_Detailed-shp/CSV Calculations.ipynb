```python
#CALCULATING CUMULATIVE OIL PRODUCTION

for i in np.arange(2001,2020): 
    
    prevfilename = "TX Production - " + str(i-1) + ".csv"    
    prevprod = pd.read_csv(prevfilename)
    prevprod = prevprod.astype({"Oil (bbl)": float})
    
    currfilename = "TX Production - " + str(i) + ".csv"    
    currentprod = pd.read_csv(currfilename)
    currentprod = currentprod.astype({"Oil (bbl)": float})
    
    growth = np.arange(254, dtype=int)

    for (i,j,k) in zip(prevprod["Oil (bbl)"],currentprod["Oil (bbl)"],growth):
        if i == 0:
            growth[k] = 0
        
        else:
            growth[k] = ((j-i)/i)*100
    
    currentprod["Growth Rate (%)"] = growth
    
    currentprod.to_csv(currfilename)
```
