```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
prod2000 = pd.read_csv("TX Production - 2000.csv")
oilprod = prod2000["Oil (Million BBL)"]
```

```python
fig, ax = plt.subplots(figsize=(4, 2.5), dpi=144)
colors = plt.cm.Dark2(range(6))
y = oilprod.index
width = oilprod.values
ax.barh(y=y, width=width, color=colors)
```

```python
#FIX AXIS TO MAKE PRETTIER
def nice_axes(ax):
    ax.set_facecolor('.8')
    ax.tick_params(labelsize=8, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
    [spine.set_visible(False) for spine in ax.spines.values()]
    
nice_axes(ax)
fig
```

```python
#ORDER BARS FROM LARGEST TO SMALLEST
fig, ax_array = plt.subplots(nrows=1, ncols=3, figsize=(7, 2.5), dpi=144, tight_layout=True)

years = np.arange(2001,2004)

for ax, date in zip(ax_array, years):
    
    filename = "TX Production - " + str(date) + ".csv"
    prod = pd.read_csv(filename)
    oilprod = prod["Oil (Million BBL)"]
    
    s = oilprod.sort_values()
    ax.barh(y=s.index, width=s.values, color=colors)
    ax.set_title(date, fontsize='smaller')
    nice_axes(ax)
```
