```python
import os                                                   # to set current working directory 
import numpy as np                                          # arrays and matrix math
import pandas as pd                                         # DataFrames
import matplotlib.pyplot as plt                             # for plotting
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_excel('2010_Data.xls')
df = pd.DataFrame(df)
df = df.set_index(['Year'])
original_columns = df.columns
df.head()
#df.describe()
```

```python
for i in range(len(original_columns)):
    #Year Columns
    if i%2 == 0:
        row = df[df[original_columns[i+1]] == df[original_columns[i+1]].values.max()].index[0]
        peak_year = df[original_columns[i]][row]
        min_year = df[original_columns[i]].values.min()
        df['N_' + original_columns[i]] = (df[original_columns[i]] - min_year) / (peak_year-min_year)
    #Production Columns
    else:
        df['N_' + original_columns[i]] = (df[original_columns[i]] - df[original_columns[i]].values.min()) / (df[original_columns[i]].values.max() - df[original_columns[i]].values.min())
```

```python
norm_years = df.iloc[:,8].values
norm_prod = df.iloc[:,9].values

for i in range(10,16):
    if i%2 == 0:  
        norm_years = np.concatenate((norm_years,df.iloc[:,i].values))
    else:
        norm_prod = np.concatenate((norm_prod,df.iloc[:,i].values))
```

```python
data = {'Norm Years': norm_years, 'Norm Prod': norm_prod}
DF = pd.DataFrame(data)
DF.set_index(['Norm Years'])
```

```python
import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.arima_model import ARIMA

p = 1; d = 1; q = 2

#ARIMA_model = ARIMA(df['Production'], order=(p, d, q))
#Normalized
ARIMA_model = ARIMA(DF['Norm Prod'], order=(p, d, q))  
ARIMA_model_fit = ARIMA_model.fit(transparams = False,disp=-1) 

print(ARIMA_model_fit.summary())
```

```python
#Raw data
split = 30
train = DF.iloc[:split,1]
train.index = DF.iloc[:split,0]
test = DF.iloc[split:,1]
test.index = DF.iloc[split:,0]
```

```python
plt.subplot(321)
plt.scatter(train.index, train, color='red', label='Production Train')
plt.plot(test, color='black',linewidth = 1, label='Production Test')
plt.title('Production'); plt.legend(loc='best'), plt.xlabel('Normalized Years'), plt.ylabel('Normalized Production')
plt.subplots_adjust(left=0.0, bottom=0.0, right=5.0, top=2.0, wspace=0.2, hspace=0.2)
```

```python
# the following code taken from https://www.machinelearningplus.com/time-series/arima-model-time-series-forecasting-python/
# Fit the model to training data
model = ARIMA(train, order=(1, 1, 2))  
fitted = model.fit(disp=-1)  

# Forecast with fit model
fc, se, conf = fitted.forecast(len(test), alpha=0.05)  # 95% conf

# Convert estimate and confidence interval to Pandas series
fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf[:, 0], index=test.index)
upper_series = pd.Series(conf[:, 1], index=test.index)

# Plot the training, testing and forecast with confidence intervals
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train.iloc[:10], color='red', label='DeWitt Train')
plt.plot(train.iloc[10:20], color='red', label='Gonzales Train')
plt.plot(train.iloc[20:], color='red', label='La Salle Train')
#plt.plot(train, label='training', color = 'red',linewidth = 1)
plt.plot(test, label='Maverick actual', color = 'black',linewidth = 1)
plt.plot(fc_series, label='Maverick forecast', color = 'blue',linewidth = 1)
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.15)
plt.title('Forecast vs Actuals'); plt.xlabel('Normalized Years'), plt.ylabel('Normalized Production')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```
